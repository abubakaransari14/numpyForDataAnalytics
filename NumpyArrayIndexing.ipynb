Numpy Array Indexing

```python
import numpy as np
array1 = np.array([91, 3, 5, 7, 9])
```

Array indexes are start from 0 to the end, if we want to access the the first value/element in an array we use 0 index, index and values/elements are different things you can suppose index is position 

```python
print(f"First Value in an array: {array1[0]}")
```

```python
print(array1[0])    # prints 91 => 1st element
print(array1[2])    # prints 5 => 3rd element
print(array1[4])    # prints 9 => 4th element which is the last element in our array
```

Modify Array Elements Using Index

```python
print(array1)
```

```python
array1[0] = 21
array1[3] = 0
```

```python
print(array1)
print(len(array1))
```

```python
# array1[5] = 91 (index 5 is out of bounds for axis 0 with size 5)
```

NumPy Negative Array Indexing

as earlier i told that in array index started from 0 from left the first value having the index 0, but in negative indexing the last index is said to be -1 which means the last index is -1 and the index is decresed from right to left like -1,-2.-3 and so on


```python
# Suppose an array
array2 = [91,32,17,11,21,5]
print(f"First Element by Positive Index: {array2[0]}")
print(f"Last Element by Positive Index: {array2[len(array2)-1]}")
print(f"First Element by Negative Index: {array2[len(array2) * (-1)]}")
print(f"Last Element by Nagative Index: {array2[-1]}")
```

Modify Array Elements Using Negative Indexing

```python
print(f"Original: {array2}")
array2[-1] = 57
array2[len(array2) * (-1)] = 1
print(f"Modified: {array2}")
```

2-D NumPy Array Indexing

```python
array3 = [
#    0 1 2 3 4   => columns
    [1,2,3,4,5],    # row 0
    [3,6,7,9,4]  # row 1
]
npArr = np.array(array3)
print(f"Access value 9 by row 1 and col 3: {npArr[1][3]}") # row 1 & column 3 => i want to access 9
print(f"Approach 2: Access value 9 by row 1 and col 3: {npArr[1,3]}") # row 1 & column 3 => i want to access 9
```

Access Row or Column of 2D Array Using Indexing

```python
npArray = np.array([ 
#     columns:       0  1  2
                    [1, 3, 5],   # row 0 
                    [7, 9, 2],   # row 1
                    [4, 6, 8]    # row 2
])

print(f"Approach1: {npArray[1][1]}")
print(f"Approach2: {npArray[1,1]}")
print(f"row1: {npArray[1,:]}")
print(f"last row: {npArray[-1,:]}")
print(f"col1: {npArray[:,1]}")
print(f"last col: {npArray[:,-1]}")
```

3-D NumPy Array Indexing

```python
npArray3D = np.array([
                    [   # slice 0
        # columns        0  1  2  3  
                        [1, 2, 3, 4],    # row 0
                        [5, 6, 7, 8],    # row 1
                        [9, 10, 11, 12]  # row 2
                    ],
                     
                    [   # slice 1
        # columns        0    1   2   3
                        [13, 14, 15, 16],   # row 0 
                        [17, 18, 19, 20],   # row 1
                        [21, 22, 23, 24]    # row 2
                    ]
])
print(f"Approach1 (target: 19): {npArray3D[1][1][2]}")
print(f"Approach2 (target: 12): {npArray3D[0,2,3]}")
print(f"Approach1 Negative(target: 19): {npArray3D[-1][-2][-2]}")
print(f"Approach2 Negative (target: 12): {npArray3D[-2,-1,-1]}")
print(f"3rd column of 2nd Slice: {npArray3D[-1,:,2]}")
print(f"3rd row of 1st Slice: {npArray3D[0,2,:]}")
```
