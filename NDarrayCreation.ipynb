Create a 2-D NumPy Array

```python
import numpy as np
array1 = [
    [1,2,3,4,5,5],
    [11,221,121,12,1,1]
]
npArry = np.array(array1)
print(npArry)
print(f"Dimension: {npArry.ndim}")
print(f"Shape: {npArry.shape}")
print(f"Type: {type(npArry)}")
```

Create a 3-D NumPy Array

```python
array2 = [
    [
        [1,2,2,2,11,4],
        [0,9,4,5,21,2]
    ],
    [
        [33,12,12,4,3,11],
        [1,12,22,24,11,4],
        
    ]
]
npArray2 = np.array(array2)
print(npArray2)
print(f"Dimension: {npArray2.ndim}")
print(f"Shape: {npArray2.shape}")
print(f"Type: {type(npArray2)}")
```

Create N-D Arrays using np.zeros()

```python
npArrayZeros = np.zeros((2,2))
print("x---------FOR 2D ARRAY---------x")
print(npArrayZeros)
print(f"Dimension: {npArrayZeros.ndim}")
print(f"Shape: {npArrayZeros.shape}")
print(f"Type: {type(npArrayZeros)}")

print("x---------FOR 3D ARRAY---------x")
npArrayZeros2 = np.zeros((2,3,4))
print(npArrayZeros2)
print(f"Dimension: {npArrayZeros2.ndim}")
print(f"Shape: {npArrayZeros2.shape}")
print(f"Type: {type(npArrayZeros2)}")
```

Create N-D Array with a Specified Value

```python
npSpecified = np.full((2,2),7)
print("x---------FOR 2D ARRAY---------x")
print(npSpecified)
print(f"Dimension: {npSpecified.ndim}")
print(f"Shape: {npSpecified.shape}")
print(f"Type: {type(npSpecified)}")

print("x---------FOR 3D ARRAY---------x")
npSpecified2 = np.full((2,3,4),7)
print(npSpecified2)
print(f"Dimension: {npSpecified2.ndim}")
print(f"Shape: {npSpecified2.shape}")
print(f"Type: {type(npSpecified2)}")
```

Creating Arrays With np.random.rand()

```python
npRandomRand = np.random.rand(2,2)
print("x---------FOR 2D ARRAY---------x")
print(npRandomRand)
print(f"Dimension: {npRandomRand.ndim}")
print(f"Shape: {npRandomRand.shape}")
print(f"Type: {type(npRandomRand)}")

print("x---------FOR 3D ARRAY---------x")
npRandomRand2 = np.random.rand(2,3,4)
print(npRandomRand2)
print(f"Dimension: {npRandomRand2.ndim}")
print(f"Shape: {npRandomRand2.shape}")
print(f"Type: {type(npRandomRand2)}")
```
