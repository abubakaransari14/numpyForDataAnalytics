Attributes are the properties of an array which define the shape, size, dimension, data type etc of an array

```python
import numpy as np
array1 = [
    [1,2,3,4,5,5],
    [11,221,121,12,1,1]
]
npArry = np.array(array1)
print(npArry)
print(f"Dimension: {npArry.ndim}")
print(f"Size (no. of element in an array): {npArry.size}")
print(f"Item Size (Size of each element): {npArry.itemsize}")
print(f"Shape: {npArry.shape}")
print(f"Data (location in a memory): {npArry.data}")
print(f"Type: {type(npArry)}")
```
