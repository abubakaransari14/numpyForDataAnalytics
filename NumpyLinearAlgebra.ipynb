NumPy Linear Algebra Functions

NumPy dot() Function

```python
import numpy as np

array1 = np.array([1, 3, 5])
array2 = np.array([2, 4, 6])
result = np.dot(array1, array2)

print(result)
```

NumPy inner() Function

```python
array1 = np.array([[1, 3], 
                   [5, 7]])
array2 = np.array([[2, 4], 
                   [6, 8]])
result = np.inner(array1, array2)
print(result)
```

NumPy outer() Function

```python
import numpy as np

array1 = np.array([1, 3, 5])
array2 = np.array([2, 4, 6])
result = np.outer(array1, array2)

print(result)
```

NumPy det() Function

```python
array1 = np.array([[1, 3], 
                  [5, 7]])
result = np.linalg.det(array1)
print(result)
```

NumPy trace() Function

```python
array1 = np.array([[5, 3, 5], 
                   [9, 2, 1], 
                   [7, 8, 10]])
result = np.trace(array1)
print(result)
```

NumPy inv() Function

```python
import numpy as np

array1 = np.array([[2, 4], 
                  [6, 8]])
result = np.linalg.inv(array1)
print(result)
```
