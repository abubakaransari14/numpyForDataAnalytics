Numpy Random

```python
import numpy as np

randomNumber = np.random.randint(0, 10)

print(randomNumber)
```

```python
randomFloat = np.random.rand()
print(randomFloat)
```

```python

array1D = np.random.randint(0, 10 , 5)
array2D = np.random.randint(0, 10 , size=(2,2))

print("1D Random Integer Array:\n",array1D)
print("2D Random Integer Array:\n",array2D)

floatArray = np.random.rand(4) #.reshape(2,2) for 2D
print(floatArray)

# generate 2D array of shape (3, 4) with random integers
result = np.random.randint(0, 10, (3,4))
print("\n2D Random Integer Array:\n",result)
```

Choose Random Number from NumPy Array

```python
# create an array of integers from 1 to 5
array1 = np.array([1, 2, 3, 4, 5])
randomChoice = np.random.choice(array1)

print(randomChoice)

```
