NumPy Array Slicing

1D NumPy Array Slicing

```python
import numpy as np

array1 = np.array([1, 3, 5, 7, 8, 9, 2, 4, 6])
print(f"Intially till 1 to 4 element: {array1[0:4]}") #last one is ignored
print(f"Intially till 4 to last element: {array1[3:len(array1)]} ")
```

```python
array1
```

Using start Parameter

```python
print(array1[5:])
```

```python
array1[5:] = 7
print(array1)
```

```python
array2 = np.array([1, 3, 5, 17, 18, 19, 12, 14, 6])
```

Using stop Parameter

```python
print(array2[:5])
```

```python
array2[: 5] = 7
```

```python
print(array2)
```

```python
array3 = np.array([11, 31, 15, 71, 181, 191, 102, 14, 6])
```

Using start and stop parameter

```python
print(array3[2:5])
```

```python
array3[2:5] = 7
print(array3)
```

Using start, stop, and step parameter

```python
array4 = np.array([2, 4, 6, 8, 10, 12])

array4[1:5:2] = 16
print(array4)
```

NumPy Array Negative Slicing

```python

numbers = np.array([2, 4, 6, 8, 10, 12])

print(numbers[-3:])    # [8 10 12]

print(numbers[-5:-2])    # [4 6 8] 

print(numbers[-1::-2])   # [12 8 4]
```

Reverse NumPy Array Using Negative Slicing

```python

numbers = np.array([2, 4, 6, 8, 10, 12])

reversed_numbers = numbers[::-1]
print(reversed_numbers)

# Output: [12 10 8 6 4 2]
```

2D NumPy Array Slicing

```python
# create a 2D array
npArr2D = np.array([[1, 3, 5, 7], 
                   [9, 11, 13, 15],
                   [1, 3, 5, 7], 
                   [9, 11, 13, 15]])
print(npArr2D)
```

```python
print(npArr2D[:,3]) #columns and row
```

```python
npArr2D[0:2,0:2]
```

```python
npArr2D[:2,:2]
```

```python
npArr2D[0:4:2,0:4:2]
```
