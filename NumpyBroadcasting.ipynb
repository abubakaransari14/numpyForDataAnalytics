Advance Numpy Operations

```python
import numpy as np


array1 = np.array([1, 2, 3])
array2 = np.array([[1], [2], [3]])

sum = array1 + array2

print(sum)
```

Compatibility Rules for Broadcasting

```python

array1 = np.array([1, 2, 3])
array2 = np.array([[1,3,4], 
                   [2,5,5], 
                   [1,4,3],
                   [1,4,3]
                  ])
number = 5

sum = array1 + number
print(array2.shape)
print(array1.shape)
print(array1 + array2)
print(sum)
```
