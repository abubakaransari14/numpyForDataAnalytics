```python
import numpy as np

array1 = np.array([1, 3, 5, 7])
array2 = np.array([2, 4, 6, 8])
print(f"Original Array: {array1} and {array2}")
addArr = array1 + array2
print("Using the + operator:",addArr) 

addArr2 = np.add(array1, array2)
print("Using the add() function:",addArr2) 
```

NumPy Array Element-Wise Subtraction

```python
sub = array1 - array2
print("Using the - operator:",sub) 

sub2 = np.subtract(array1, array2)
print("Using the subtract() function:",sub) 
```

NumPy Array Element-Wise Multiplication

```python
mult = array1 * array2
print("Using the * operator:",mult) 

mult2 = np.multiply(array1, array2)
print("Using the multiply() function:",mult2) 
```

NumPy Array Element-Wise Division

```python
dvd = array1 / array2
print("Using the / operator:",dvd) 

dvd2 = np.divide(array1, array2)
print("Using the divide() function:",dvd2) 
```

NumPy Array Element-Wise Exponentiation

```python
exp = array1 ** 2
print("Using the ** operator:",exp) 

exp2 = np.power(array1,2)
print("Using the power() function:",exp2) 
```

NumPy Array Element-Wise Modulus

```python
array3 = np.array([9, 10, 20])
array4 = np.array([2, 5, 7])
mod = array3 % array4
print("Using the % operator:",mod) 

mod2 = np.mod(array3,array4)
print("Using the mod() function:",mod2) 
```
