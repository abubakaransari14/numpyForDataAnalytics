NumPy Histogram

```python
import numpy as np

data = np.array([5, 10, 15, 18, 20])

bin = [0,10,20,30]

graph = np.histogram(data, bin)

print(graph)
```

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.array([5, 10, 15, 18, 20])
bins = [0,10,20,30]

graph = np.histogram(data, bins)
print(graph)
plt.hist(data, bins)

plt.show()
```
