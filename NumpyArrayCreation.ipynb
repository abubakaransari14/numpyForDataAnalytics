Numpy is a python library created at 2005 that perform numericl calculation, it is generally used for working with an array.

Why we use numpy?

In python we use list to work with the array but when it comes to large operation python list are not enough optimized. Numpy Array are 50X faster than a python list and its used in complex mathematical and statistical operation

```python
import numpy as np
```

```python
array1 = [1,2,3,44,5,1]
print(type(array1))
```

```python
npArray = np.array(array1)
print(type(npArray))
```

Create an Array Using np.zeros()

```python
npZeros = np.zeros(5)
print(npZeros)
```

```python
npZeros.ndim
```

```python
npZeros2 = np.zeros((2,3))
print(npZeros2)
print(f"Dimension: {npZeros.ndim}")
```

Create an Array With np.arange()

```python
npArange = np.arange(10)
print(npArange)
print(f"Dimension: {npArange.ndim}")
```

```python
npArange = np.arange(10).reshape(5,2)
print(npArange)
print(f"Dimension: {npArange.ndim}")
print(f"Dimension: {npArange.shape}")
```

Create an Array With np.random.rand()

```python
npRandomRand = np.random.rand(5)
print(npRandomRand)
print(f"Dimension: {npRandomRand.ndim}")
```

```python
npRandomRand = np.random.rand(2,3)
print(npRandomRand)
print(f"Dimension: {npRandomRand.ndim}")
```

Create an Empty NumPy Array

```python
npEmpty = np.empty(4)
print(npEmpty)
print(f"Dimension: {npEmpty.ndim}")
```

```python
npEmpty = np.empty((2,3))
print(npEmpty)
print(f"Dimension: {npEmpty.ndim}")
```
