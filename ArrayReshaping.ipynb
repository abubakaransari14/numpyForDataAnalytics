NumPy array reshaping simply means changing the shape of an array without changing its data.

```python
import numpy as np
```

```python
array1 = np.array([1, 3, 5, 7, 2, 4, 6, 8])
```

```python
print(array1)
```

syntax: np.reshape(array, newshape, order = 'C')

```python
newArr = np.reshape(array1, (2,2,2), order= 'C')
```

```python
print(newArr)
```

```python
newArr = np.reshape(array1, (2,4), order= 'C')
```

```python
print(newArr)
```

```python

array2 = np.array([[1, 3], [5, 7], [9, 11]])
result2 = np.reshape(array2, -1)
print("Flattened 2D array:", result2)

# flatten 3D array to 1D
array3 = np.array([[[1, 3], [5, 7]], [[2, 4], [6, 8]]])
result3 = np.reshape(array3, -1)
print("Flattened 3D array:", result3)
```
